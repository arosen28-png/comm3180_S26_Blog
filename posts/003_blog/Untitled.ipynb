# What Our Personalities Say About Us

For this post, I explored several personality datasets to see what patterns might show up. The data includes a massive personality dataset with nearly 20,000 people, essays written by thousands of participants, MBTI personality types from social media posts, and even the personality scores from our own class.

The goal was simple: see what the data might reveal about how people think, act, and express themselves.


## Looking at the Bigger Picture

The first dataset contains personality survey results from almost **20,000 people**. It measures the five major personality traits:

* Extraversion
* Neuroticism
* Agreeableness
* Conscientiousness
* Openness

With such a large dataset, we can start to see what personality looks like across a broad group of people. Most people tend to fall somewhere in the middle on these traits. Very extreme personalities are actually pretty rare.

This suggests that while personality differences exist, most people are not completely opposite from each other. Instead, personality tends to sit on a spectrum where people cluster near the center.


## How Does Our Class Compare?

One of the most interesting parts of the analysis was comparing the **large personality dataset** with the results from our own class.

Our class dataset is much smaller, but it still shows a few interesting patterns. Some students scored very high in extraversion, while others scored much lower. The range of scores suggests that even in a relatively small group, personalities can vary a lot.

Compared to the larger dataset, our class seems to have a fairly wide mix of personality types, which is probably what you would expect from a group of students with different backgrounds and interests.


## Personality and Writing

Another dataset included **over 2,000 personal essays** written by people who also took personality tests.

This dataset is interesting because it connects personality to how people write. For example, people who scored high in extraversion often wrote in a more energetic and social way, while people higher in neuroticism sometimes used more emotional language.

It shows that personality can actually show up in subtle ways in the words people choose.


## Personality on Social Media

The final dataset looked at **MBTI personality types** alongside thousands of social media posts.

Unlike the Big Five, MBTI sorts people into specific personality categories like INTJ or ENFP. When looking at the posts, you can sometimes see differences in tone or topics depending on the personality type.

For example, some personality types tend to write longer or more reflective posts, while others post more casually or socially.


## What This All Suggests

Looking across these datasets, one thing becomes clear: personality is complicated.

No single test captures everything about a person. But when you look at personality from multiple angles — surveys, essays, and social media — interesting patterns start to appear.

The data suggests that personality influences how people think, how they write, and even how they communicate online.

At the same time, most people are not extremely different from each other. Instead, personality tends to exist on a spectrum where people share many traits but express them in slightly different ways.

In the end, personality tests might not perfectly define who we are, but they can still reveal some surprising insights about human behavior.
