```python
import json
import pandas as pd
import matplotlib.pyplot as plt

with open("bicycle-crash-data-chapel-hill-region.json", "r") as f:
    data = json.load(f)

df = pd.DataFrame(data)

df["sev_simple"] = df["crsh_sevri"].str[0]

light_counts = df["light_cond"].value_counts()

plt.figure(figsize=(10,5))
light_counts.plot(kind="bar")
plt.title("Bike Crashes by Light Condition")
plt.xlabel("Light condition")
plt.ylabel("Crash count")
plt.tight_layout()
plt.show()
```

```python
serious = df[df["sev_simple"].isin(["A", "K"])].copy()

serious_light = serious["light_cond"].value_counts()

plt.figure(figsize=(10,5))
serious_light.plot(kind="bar")
plt.title("Most Severe Bike Crashes by Light Condition")
plt.xlabel("Light condition")
plt.ylabel("Serious crash count")
plt.tight_layout()
plt.show()
```

```python
speed_counts = df["speed_limi"].value_counts().sort_index()

plt.figure(figsize=(10,5))
speed_counts.plot(kind="bar")
plt.title("Bike Crashes by Posted Speed Limit")
plt.xlabel("Speed limit")
plt.ylabel("Crash count")
plt.tight_layout()
plt.show()
```

```python
city_counts = df["city"].value_counts()

plt.figure(figsize=(10,5))
city_counts.plot(kind="bar")
plt.title("Bike Crashes by City")
plt.xlabel("City")
plt.ylabel("Crash count")
plt.tight_layout()
plt.show()
```
